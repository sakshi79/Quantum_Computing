```python
circuit = QuantumCircuit(6+1, 6)
circuit.h([0,1,2,3,4,5])
circuit.x(6)
circuit.h(6)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
results = execute(circuit, backend=simulator, shots=1).result()
counts = results.get_counts()
print(counts)
```
